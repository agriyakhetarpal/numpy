```python
import numpy as np
x = np.ma.array(np.arange(5), mask=[0]*2 + [1]*3)
x.compressed()
```

N-D arrays are compressed to 1-D.


```python
arr = [[1, 2], [3, 4]]
mask = [[1, 0], [0, 1]]
x = np.ma.array(arr, mask=mask)
x.compressed()
```
